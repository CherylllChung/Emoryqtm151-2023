# percentage of orders with issues reported accross ratings

```python

mysql = lambda q: sqldf(q, globals())
mysql("select customer_order_rating, count(distinct case when type_of_issue_reported is not null then order_id end)，count(distinct order_id) from df group by 1 ;")
```

# distribution of orders with rating < 4 at different regions

```python
#oberservation 2，number of orders with rating 〈= 4 in different region


mysql = lambda q: sqldf(q, globals())
mysql("select region, count(distinct case when customer_order_rating < 4 then order_id else 0 end),count(distinct order_id) from df group by 1 ;")
```

# daily trend of number of orders and number of orders with issue in sf

```python
# obeservation 4，daily trend of number of orders with no issue and issue  in sf
mysql = lambda q: sqldf(q, globals())
mysql("select substring(order_delivery_time,1,10) as day, count(distinct order_id) as order_ct, count(distinct case when TYPE_OF_ISSUE_REPORTED is not null then order_id end)from df where region = 'sf' group by 1 order by 1;")

```

# hourly trend of number of orders and number of orders with issue  in sf

```python
# obeservation 4，hourly trend of number of orders and number of orders with issue in sf
mysql = lambda q: sqldf(q, globals())
mysql("select substring(order_delivery_time,12,2) as hour, count(distinct order_id) as order_ct, sum(case when TYPE_OF_ISSUE_REPORTED is not null then 1 else 0 end)from df where region = 'sf' group by 1 order by 1;")


```
